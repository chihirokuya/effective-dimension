Here, we report the results from pynotes/neural_netowrk/local.ipynb. gen_error is computed simply as abs(test_error - train_error)

```python
import matplotlib.pyplot as plt
import sys
sys.path.append('../../')

from package.plot_utils.constant import DPI, TWO_PLOTS_FIGSIZE
```

```python
def plot(data: dict):
    eff_dims = []
    props = []
    generalization_erros = []
    for key, value in data.items():
        props.append(value['test_error'] / value['train_error'])
        generalization_erros.append(value['gen_error'])
        eff_dims.append(value['eff_dim'])

    fig, axs = plt.subplots(1, 2, figsize=TWO_PLOTS_FIGSIZE, dpi=DPI)

    ax1 = axs[0]
    ax1.plot(range(6), eff_dims, 'b-o')
    ax1.set_ylabel('Normalized Local Effective Dimension', color='b')
    ax1.tick_params(axis='y', labelcolor='b')
    ax2 = ax1.twinx()
    ax2.plot(range(6), generalization_erros, 'r-o')
    ax2.set_ylabel('Test Loss - Train Loss', color='r')
    ax2.tick_params(axis='y', labelcolor='r')
    ax1.set_xticks(range(6), [round(0.1*i, 1) for i in range(6)])
    ax1.set_xlabel('Dropout Rate')


    ax1 = axs[1]
    ax1.plot(range(6), eff_dims, 'b-o')
    ax1.set_ylabel('Normalized Local Effective Dimension', color='b')
    ax1.tick_params(axis='y', labelcolor='b')
    ax2 = ax1.twinx()
    ax2.plot(range(6), props, 'r-o')
    ax2.set_ylabel('Test Loss / Train Loss', color='r')
    ax2.tick_params(axis='y', labelcolor='r')
    ax1.set_xticks(range(6), [round(0.1*i, 1) for i in range(6)])
    ax1.set_xlabel('Dropout Rate')

    fig.tight_layout(pad=3)
    plt.savefig('gen_3')
```
